```python
import turtle

wn = turtle.Screen()
wn.title("Tennis")
wn.bgcolor("green")  # Tennis courts are often green
wn.setup(width=900, height=700)
wn.tracer(0)

# Score
score_player1 = 0
score_player2 = 0

# Racket A
racket_a = turtle.Turtle()
racket_a.speed(0)
racket_a.shape("square")
racket_a.color("white")
racket_a.shapesize(stretch_wid=5, stretch_len=1)  # Adjust size to resemble a tennis racket
racket_a.penup()
racket_a.goto(-350, 0)

# Racket B
racket_b = turtle.Turtle()
racket_b.speed(0)
racket_b.shape("square")
racket_b.color("white")
racket_b.shapesize(stretch_wid=5, stretch_len=1)  # Adjust size to resemble a tennis racket
racket_b.penup()
racket_b.goto(350, 0)

# Tennis Ball
tennis_ball = turtle.Turtle()
tennis_ball.speed(0)
tennis_ball.shape("circle")  # Change shape to circle to resemble a tennis ball
tennis_ball.color("yellow")  # Tennis balls are usually yellow
tennis_ball.penup()
tennis_ball.goto(0, 0)
tennis_ball.dx = 3
tennis_ball.dy = -3

# Scoreboard
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 260)
pen.write("Player 1: 0   Player 2: 0", align="center", font=("Courier", 24, "normal"))
```

```python
# Function to move rackets
def racket_a_up():
    y = racket_a.ycor()
    y += 20
    racket_a.sety(y)

def racket_a_down():
    y = racket_a.ycor()
    y -= 20
    racket_a.sety(y)

def racket_b_up():
    y = racket_b.ycor()
    y += 20
    racket_b.sety(y)

def racket_b_down():
    y = racket_b.ycor()
    y -= 20
    racket_b.sety(y)

# Keyboard bindings
wn.listen()
wn.onkeypress(racket_a_up, "w")
wn.onkeypress(racket_a_down, "s")
wn.onkeypress(racket_b_up, "Up")
wn.onkeypress(racket_b_down, "Down")
```

```python
# Main game loop
while True:
    wn.update()
    
    # Move the ball
    tennis_ball.setx(tennis_ball.xcor() + tennis_ball.dx)
    tennis_ball.sety(tennis_ball.ycor() + tennis_ball.dy)
    
    # Border checking
    if tennis_ball.ycor() > 290:
        tennis_ball.sety(290)
        tennis_ball.dy *= -1
        
    if tennis_ball.ycor() < -290:
        tennis_ball.sety(-290)
        tennis_ball.dy *= -1
        
    if tennis_ball.xcor() > 390:
        tennis_ball.goto(0, 0)
        tennis_ball.dx *= -1
        score_player1 += 1
        pen.clear()
        pen.write("Player 1: {}   Player 2: {}".format(score_player1, score_player2), align="center", font=("Courier", 24, "normal"))
        
    if tennis_ball.xcor() < -390:
        tennis_ball.goto(0, 0)
        tennis_ball.dx *= -1
        score_player2 += 1
        pen.clear()
        pen.write("Player 1: {}   Player 2: {}".format(score_player1, score_player2), align="center", font=("Courier", 24, "normal"))
        
    # Racket and ball collisions
    if (tennis_ball.xcor() > 340 and tennis_ball.xcor() < 350) and (tennis_ball.ycor() < racket_b.ycor() + 50 and tennis_ball.ycor() > racket_b.ycor() - 50):
        tennis_ball.setx(340)
        tennis_ball.dx *= -1
```
